```
import pandas as pd

from May_2025.Day7_22_05_2025.ML_data_wrangling_part_1 import label_object

df = pd.read_csv("../../Dataset/Diamonds/diamonds.csv")
df.head(10)
```

```
from sklearn.preprocessing import LabelEncoder

label_object_LE = LabelEncoder()
```

```
df['cut'] = label_object_LE.fit_transform(df['cut'])
df['color'] = label_object_LE.fit_transform(df['color'])
df['clarity'] = label_object_LE.fit_transform(df['clarity'])
df.head(20)
```

```
from sklearn.preprocessing import OrdinalEncoder

label_object_OE = OrdinalEncoder()
```

```
df['cut'] = label_object_OE.fit_transform(df[['cut']])
df['color'] = label_object_OE.fit_transform(df[['color']])
df['clarity'] = label_object_OE.fit_transform(df[['clarity']])
df.head(20)
```
